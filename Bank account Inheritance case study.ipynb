```python
# WITH INHERITANCE

class gitBank:
    def __init__(self,acc_type,acc_num,cust_fname,cust_lname,bal_amount):
        self.acc_type=acc_type
        self.acc_num=acc_num
        self.cust_fname=cust_fname
        self.cust_lname=cust_lname
        self.bal_amount=bal_amount
    def balenquiry(self):
        print('\t\t\t\t---------------------------')
        print('\t\t\t\tAccount Type :',self.acc_type)
        print('\t\t\t\tAccount Number :',self.acc_num)
        print('\t\t\t\tBalance Amount : ',self.bal_amount,'/-')
        print('\t\t\t\t---------------------------')
    def deposit(self,amount):
        self.amount=amount
        self.bal_amount += amount
        print('\t\t\t\t-------------------------------')
        print('\t\t\t\tAmount deposited in ',self.acc_num,'---',self.amount,"/-")
        print('\t\t\t\tUpdated Balance -- ',self.bal_amount,"/-")
        print('\t\t\t\t-------------------------------')
    def withdraw(self,amount):
        self.amount=amount
        if self.amount<self.bal_amount:
            self.bal_amount=self.bal_amount-amount
            print('\t\t\t\t--------------------------------')
            print('\t\t\t\tAmount withdrawn in ',self.acc_num,'---',self.amount,"/-")
            print("\t\t\t\tRemaining balance in your account :",self.bal_amount,"/-")
            print('\t\t\t\t--------------------------------')
        else:
            print("\t\t\t\tInsufficient Funds in your account\n")
    def transfer(self,acc_name):
        # taking input from user for amount to transfer
        amt=int(input("Enter the amount to transfer :"))
        if amt<self.bal_amount:
            self.withdraw(amt)
            acc_name.deposit(amt)   
        else:
            # if transfer amount is more than balance loop runs again
            print("\t\t\t\tInsufficient Funds in your account\n")
            amt=int(input("Enter the amount to transfer :"))
            if amt<self.bal_amount:
                self.withdraw(amt)
                acc_name.deposit(amt)
            else:
                print("\t\t\t\tInsufficient Funds in your account\n")


class savingsAccount(gitBank):
    def __init__(self,acc_num,cust_fname,cust_lname,bal_amount):
        gitBank.__init__(self,'Savings Account',acc_num,cust_fname,cust_lname,bal_amount)
        print('\t\t-------------WELCOME TO Git Bank -----------\n')
        print("\t\t\tYour SAVINGS ACCOUNT is created and is active")
        print('\t\t\tYou can start using our services\n')
    def feature(self):
        print("Interest Rate: 6-8%")
        print("All types of Loans available")
        print("Gold Loan available at no interest")
    def offers(self):
        print("Online Shopping get 5% cashback")
        print("All types of tavel Ticket booking get 8 % cashback")
        print("Get every month 2 movie tickets free")
        

        
class currentAccount(gitBank):
    def __init__(self,acc_num,cust_fname,cust_lname,bal_amount):
        gitBank.__init__(self,'Current Account',acc_num,cust_fname,cust_lname,bal_amount)
        print('\t\t------------- WELCOME TO Git Bank -----------\n')
        print("\t\t\tYour CURRENT ACCOUNT is created and is active")
        print('\t\t\tyou can start using our services\n')
    def feature(self):
        print("Interest Rate: 7%")
        print("No Loans available")
        print("Gold Loan available at no interest")
    def offers(self):
        print("Online Shopping get 10% cashback")
        print("All types of tavel Ticket booking get 10 % cashback")
        print("Get every month 2 movie tickets free")
        
IBSA00001=savingsAccount(919494393925,'Chirstopher','Nolan',50000)
IBCA00001=currentAccount(20220690980,'Superman','Batman',100000)
IBCA00002=currentAccount(20220690102,'Spiderman','Wonderwomen',2050000)

IBSA00001.deposit(3000)
IBCA00002.withdraw(50000)
IBCA00001.balenquiry()
IBCA00001.transfer(IBCA00002)
```
